```python
import numpy as np
import pandas as pd  
import matplotlib.pyplot as plt 

df=pd.read_csv("D:/machine_learning_code/datasets_for_models/Social_Network_Ads.csv")
x=df.iloc[:,:-1].values #assiging the first 2 columns of the dataset as they are dependent vaiables 
y=df.iloc[:,-1].values #assiging the last column (target variable) as it is independent variable and that is column we are going to predict.
# there is no rule to take only the last column as the target var, it dependend on the problem statement.
df.shape #to know the number of row and colmns of the dataset, here there are 400 rows and 3 columns.
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

```

```python
print(x_train)
```

```python
print(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(x_train,y_train)
```

```python
# now the model is trained, next step is to make the predictions on the test data
y_pred=classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
# to predict a new single value
print(classifier.predict(sc.transform([[30,87000]])))
```

```python
# evaluate the model
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
ac=accuracy_score(y_test, y_pred)
print(f'accuracy: {ac} in %: {ac*100}')
```
