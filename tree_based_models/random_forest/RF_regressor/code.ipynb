```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("D:/machine_learning_code/datasets_for_models/Position_Salaries.xlsx")
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

```python
df.head()
```

```python
df.tail()
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=3,random_state=0)
model=regressor.fit(x,y)
```

```python
regressor.predict([[6.5]])
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
y_pred=model.predict(x)
print(f"R2 Score: {r2_score(y, y_pred):.2f}")
print(f"Mean Squared Error: {mean_squared_error(y, y_pred):.2f}")
print(f"Mean Absolute Error: {mean_absolute_error(y, y_pred):.2f}")
```
